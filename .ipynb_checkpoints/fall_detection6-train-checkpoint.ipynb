```python
# Testar o YOLO - Fazer a predição em um vídeo
# Abre o vídeo
VIDEO_PATH = "videos_aug" # _le2i
FILENAME_IN = "002_aug7_combo1"
FILENAME_OUT = "002_aug7_combo1A"
cap = cv2.VideoCapture(f"{VIDEO_PATH}/{FILENAME_IN}.mp4")

# Verifica se abriu
if not cap.isOpened():
    print("Erro ao abrir o vídeo!")
else:
    print("Vídeo carregado com sucesso!")

# Informações do vídeo
ret, frame = cap.read()
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"Resolução: {frame_width}x{frame_height}, FPS: {fps}")
plt.figure(figsize=(4,2))
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Saída de vídeo
# fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec para .mp4
# out = cv2.VideoWriter('001A.mp4', fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))  # Ajuste a resolução
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # XVID
output_path = f"{VIDEO_PATH}/{FILENAME_OUT}.mp4"
out = cv2.VideoWriter(output_path, fourcc, 20.0, (frame_width, frame_height))
loading = "Loading..."
frames_processed = 0
dot_count = 1

print(f"Salvando em: {output_path}")
print(f"Diretório existe? {os.path.isdir(VIDEO_PATH)}")
print(f"Resolução: {frame_width},{frame_height}")

if not out.isOpened():
    print("[ERRO] Falha ao inicializar VideoWriter.")
```

```python
# BOX
# Loop de processamento de frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Predição com o modelo YOLO
    print(f'\r{loading}', end='', flush=True)
    
    results = yolo_model.predict(
        source=frame, 
        imgsz=frame_width, 
        conf=0.5, 
        save=False, 
        verbose=False
    )
    #annotated_frame = results[0].plot()  # Frame com as anotações do YOLO
    #annotated_frame_resized = cv2.resize(annotated_frame, (frame_width, frame_height))

    class_info = {
        0: {"name": "no_fall", "color": (255, 0, 0)},     # Azul
        1: {"name": "fall", "color": (0, 0, 255)},        # Vermelho
        2: {"name": "attention", "color": (0, 255, 255)}  # Amarelo
    }

    # Clonar frame para desenhar manualmente
    annotated_frame = frame.copy()
    
    # Verifica se há detecções
    if results[0].keypoints is not None and results[0].boxes is not None:
        if len(results[0].boxes.data) > 0:
            boxes = results[0].boxes.xyxy.cpu().numpy()
            classes = results[0].boxes.cls.cpu().numpy().astype(int)
            confs = results[0].boxes.conf.cpu().numpy()
    
            #if keypoints.shape[0] > 0:  # existe pelo menos 1 frame
            for i, (box, cls_id, conf) in enumerate(zip(boxes, classes, confs)):
                cls_data = class_info.get(cls_id)
                color = cls_data["color"]
                cls_name = cls_data["name"]
                
                # Desenha bounding box
                x1, y1, x2, y2 = map(int, box)
                cv2.rectangle(annotated_frame, (x1, y1), (x2, y2), color, 2)
    
                # Classe e confiança
                label = f"{cls_name} ({conf:.2f})"
                cv2.putText(annotated_frame, label, (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
    
    annotated_frame_resized = cv2.resize(annotated_frame, (frame_width, frame_height))

    # Escreve o frame anotado no arquivo de saída
    out.write(annotated_frame_resized)
    frames_processed += 1

# Libera os objetos de captura e escrita de vídeo
cap.release()
out.release()

print(f"{frames_processed} frames processados e salvos em {output_path}")
```

```python
# POSE
# Loop de processamento de frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Predição com o modelo YOLO
    print(f'\r{loading}', end='', flush=True)
    
    results = yolo_model.predict(
        source=frame, 
        imgsz=frame_width, 
        conf=0.5, 
        save=False, 
        verbose=False
    )
    #annotated_frame = results[0].plot()  # Frame com as anotações do YOLO
    #annotated_frame_resized = cv2.resize(annotated_frame, (frame_width, frame_height))

    class_info = {
        0: {"name": "no_fall", "color": (255, 0, 0)},     # Azul
        1: {"name": "fall", "color": (0, 0, 255)},        # Vermelho
        2: {"name": "attention", "color": (0, 255, 255)}  # Amarelo
    }

    skeleton_connections = [
        (6, 8), (8, 10),      # Braço esquerdo
        (7, 9), (9, 11),      # Braço direito
        (12, 14), (14, 16),   # Perna esquerda
        (13, 15), (15, 17),   # Perna direita
        (5, 6), (5, 7),       # Ombros
        (12, 13),             # Quadris
        (6, 12), (7, 13),     # Tronco
        (0, 1), (1, 3), (0, 2), (2, 4),  # Cabeça (se houver)
        (0, 5),               # Pescoço → ombros
    ]

    # Clonar frame para desenhar manualmente
    annotated_frame = frame.copy()
    
    # Verifica se há detecções
    # if results[0].keypoints is not None and results[0].boxes is not None:
    if len(results[0].keypoints.data) > 0:
        keypoints = results[0].keypoints.data.cpu().numpy()
        boxes = results[0].boxes.xyxy.cpu().numpy()
        classes = results[0].boxes.cls.cpu().numpy().astype(int)
        confs = results[0].boxes.conf.cpu().numpy()

        if keypoints.shape[0] > 0:  # existe pelo menos 1 pessoa
            for i, (box, cls_id, conf, person_kps) in enumerate(zip(boxes, classes, confs, keypoints)):
                cls_data = class_info.get(cls_id)
                color = cls_data["color"]
                cls_name = cls_data["name"]
                
                # Desenha bounding box
                x1, y1, x2, y2 = map(int, box)
                cv2.rectangle(annotated_frame, (x1, y1), (x2, y2), color, 2)

                # Classe e confiança
                label = f"{cls_name} ({conf:.2f})"
                cv2.putText(annotated_frame, label, (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

                # Pontos
                for x, y, c in person_kps:
                    if c > 0.3:
                        cv2.circle(annotated_frame, (int(x), int(y)), 3, color, -1)

                # Conexões
                for start, end in skeleton_connections:
                    if start < len(person_kps) and end < len(person_kps):
                        if person_kps[start][2] > 0.3 and person_kps[end][2] > 0.3:
                            pt1 = tuple(person_kps[start][:2].astype(int))
                            pt2 = tuple(person_kps[end][:2].astype(int))
                            cv2.line(annotated_frame, pt1, pt2, (0, 255, 255), 2)
    
    annotated_frame_resized = cv2.resize(annotated_frame, (frame_width, frame_height))

    # Escreve o frame anotado no arquivo de saída
    out.write(annotated_frame_resized)
    frames_processed += 1

# Libera os objetos de captura e escrita de vídeo
cap.release()
out.release()

print(f"{frames_processed} frames processados e salvos em {output_path}")
```

```python
def extract_pose_features(pose_data, frame_shape):
    keypoints = []
    frame_height, frame_width = frame_shape[:2]
    
    if pose_data is None or len(pose_data) == 0:
        return np.zeros(54, dtype=np.float32)
    
    for x, y, c in pose_data:
        keypoints.extend([x / frame_width, y / frame_height, c])
    return np.array(keypoints, dtype=np.float32)
```

```python
def process_video(video_path, model):
    cap = cv2.VideoCapture(video_path)
    features = []
    labels = []  # Supondo que você extraia rótulos também
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Aqui você aplica seu modelo YOLO + Pose para obter keypoints
        # Exemplo fictício:
        pose_data = model.predict_pose(frame)  # Deve retornar [[x1, y1, c1], ..., [x18, y18, c18]]
        label = model.predict_label(pose_data)  # Por exemplo: 0 = no_fall, 1 = fall, 2 = attention

        feat = extract_pose_features(pose_data, frame.shape)
        features.append(feat)
        labels.append(label)
    
    cap.release()
    return np.array(features, dtype=np.float32), np.array(labels, dtype=np.int32)
```
