```python
### Extrair Features com YOLOv8 para o LSTM
### Script de extração de features – dados numericos que descrevem o que o YOLO viu em cada frame (as classes, medidas das caixas, e probabilidades (confidences)):

def extract_features_and_labels_from_video(video_frames_dir, yolo_model, max_boxes=10):
    features = []
    labels = []
    
    # Pega os frames ordenados
    frame_files = sorted(os.listdir(video_frames_dir))
    
    for frame_file in frame_files:
        frame_path = os.path.join(video_frames_dir, frame_file)
        label_dir = video_frames_dir.replace("images", "labels")
        label_path = os.path.join(label_dir, frame_file.replace('.PNG', '.txt'))  # Arquivo de label
        
        # Lê e converte o frame
        img = cv2.imread(frame_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Inferência YOLOv8
        result = yolo_model.predict(source=img_rgb, save=False, verbose=False)

        # Pegando apenas as bounding boxes normalizadas (x_center, y_center, width, height)
        bboxes = result[0].boxes.xywhn.cpu().numpy()  # Vetor 4 valores por detecção
        # Adicionar confidence para mostrar o grau de certeza
        confidences = result[0].boxes.conf.cpu().numpy().reshape(-1, 1)
        bboxes_with_conf = np.hstack((bboxes, confidences))
        
        # Se não detectar nada, adiciona vetor vazio (zero-padding)
        if len(bboxes_with_conf) == 0:
            bboxes_with_conf = np.zeros((max_boxes, 5))  # Cria um vetor de padding para 10 boxes, todos com zeros
        elif len(bboxes_with_conf) < max_boxes:
            padding = np.zeros((max_boxes - len(bboxes_with_conf), 5))  # Preenche com zeros
            bboxes_with_conf = np.vstack((bboxes_with_conf, padding))  # Junta as detecções com o padding
        else:
            bboxes_with_conf = bboxes_with_conf[:max_boxes]  # Se houver mais que 'max_boxes', pega apenas o número desejado

        features.append(bboxes_with_conf.flatten())  # Achata para manter o formato de vetor de 1D
        
        # Lê o arquivo de label
        # 0 = no_fall, 1 = attention, 2 = fall
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                label_line = f.readline().strip()
                if label_line:  # se a linha não estiver vazia
                    label = int(label_line.split()[0])  # pega a classe
                else:
                    label = 0  # se o arquivo estiver vazio
                labels.append(label)
        else:
            labels.append(0)  # Label 0 se não encontrar

    # Verifica se todas as features têm a mesma forma antes de empilhar
    feature_array = np.array(features)
    label_array = np.array(labels)
    
    return feature_array, label_array
```

```python
# Preparar o Dataset para LSTM
class FallDetectionDataset(Dataset):
    def __init__(self, features, labels, seq_length, transform=None):
        self.features = features
        self.labels = labels
        self.seq_length = seq_length
        self.transform = transform # Para aumentar os dados se desejar

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        seq_features = self.features[idx]  # shape (200, 50)
        seq_labels = self.labels[idx]      # shape (200,)
        
        return torch.tensor(seq_features, dtype=torch.float32), torch.tensor(seq_labels, dtype=torch.long)
        # return torch.tensor(self.features[idx], dtype=torch.float32), torch.tensor(self.labels[idx], dtype=torch.long)
```

```python
def get_videos_qty(type, vid_range):
    # Diretórios dos vídeos (frames) de validação
    video_dirs = [
        f"/home/kelson/Projects/MBA/TCC/fall_detection/experiment1/dataset-xxx/fall50/datasets/images/{type}/{vid:02d}"
        for vid in range(vid_range[0], vid_range[1])
    ]
    
    video_lengths = []
    
    # Extrair features e labels para o conjunto de validação
    for idx, video_dir in enumerate(video_dirs):
        frame_len = len(os.listdir(video_dir))
        video_lengths.append(frame_len)
    
    return video_lengths
```

```python
def prepare_data(model, type, vid_range, seq_length, step_size):
    # VALIDAÇÃO
    # Diretórios dos vídeos (frames) de validação
    video_dirs = [
        f"/home/kelson/Projects/MBA/TCC/fall_detection/experiment1/dataset-xxx/fall50/datasets/images/{type}/{vid:02d}"
        for vid in range(vid_range[0], vid_range[1])
    ]
    
    # Listas para armazenar features e labels de validação
    seq_features = []
    seq_labels = []
    video_lengths = []
    frames_len = 0
    frame_min = 1000
    frame_max = 0
    
    # Extrair features e labels para o conjunto de validação
    for idx, video_dir in enumerate(video_dirs):
        frame_len = len(os.listdir(video_dir))
        frames_len += frame_len
        frame_min = frame_len if frame_len < frame_min else frame_min
        frame_max = frame_len if frame_len > frame_max else frame_max
        video_lengths.append(frame_len)
        print(f"Processando vídeo {idx + 1}/{len(video_dirs)}... Tamanho: {frame_len} frames - Média: {frames_len/(idx + 1):.2f} frames - Menor: {frame_min} - Maior: {frame_max}")
    
        frame_features, frame_labels = extract_features_and_labels_from_video(video_dir, model)
    
        video_len = len(frame_features)
    
        # Gera as sequências de janelas deslizantes
        for i in range(0, video_len - seq_length + 1, step_size):
            window_feat = frame_features[i:i + seq_length]  # shape (SEQ_LENGTH, feature_dim)
            window_label = frame_labels[i + seq_length - 1]  # label do último frame (ou outro critério)
    
            seq_features.append(window_feat)
            seq_labels.append(window_label)
    
    print(f"Fim.")
    
    # Converte para numpy arrays no formato esperado pelo LSTM
    seq_features = np.array(seq_features)  # shape (num_sequencias, seq_length, feature_dim)
    seq_labels = np.array(seq_labels)      # shape (num_sequencias,)
    
    # Ajusta o tamanho das sequências de features
    seq_features_adjusted = np.array([adjust_sequence(seq, seq_length) for seq in seq_features])
    
    # Não é necessário ajustar os labels, pois são valores únicos por sequência
    seq_labels_adjusted = seq_labels  # Não precisa de ajuste
    
    # Normaliza as features
    seq_features_normalized = normalize_features(seq_features_adjusted)
    
    # Salva para posterior uso
    np.save(f"{type}_seq_features.npy", seq_features_normalized)
    np.save(f"{type}_seq_labels.npy", seq_labels_adjusted)
    print(f"Arquivo criado")
```

```python
def evaluate_model(loader, criterion, seq_features, seq_labels):
    # Avaliação do modelo
    lstm_model.eval()  # Coloca o modelo em modo de avaliação
    lstm_model.to(device)
    correct = 0
    total = 0
    x_loss = 0.0
    all_preds = []
    all_labels = []

    # Colocar as features de validação no formato tensor e no dispositivo correto
    seq_features_tensor = torch.tensor(seq_features, dtype=torch.float32).to(device)
    
    with torch.no_grad():  # Desativa o cálculo de gradientes
        model_predictions = lstm_model(seq_features_tensor)
        
        for batch_features, batch_labels in loader:
            outputs = lstm_model(batch_features)
    
            # Calcula a loss se quiser
            loss = criterion(outputs, batch_labels)
            x_loss += loss.item()
            
            _, predicted = torch.max(outputs, 1)
            total += batch_labels.size(0)
            correct += (predicted == batch_labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(batch_labels.cpu().numpy())
    
    accuracy = 100 * correct / total
    x_loss /= len(loader)
    
    # Caso o modelo tenha uma saída contínua (ex: probabilidade), converta para classes binárias
    model_predictions = (model_predictions > 0.5).cpu().numpy().astype(int)  # Converter para numpy se necessário
    
    y_true = seq_labels  # Rótulos reais
    y_pred = model_predictions  # Suas previsões do modelo
    # Transformando one-hot ou probabilidades em rótulos de classe
    y_pred_labels = np.argmax(y_pred, axis=1)
    
    # Cálculo de métricas
    precision = precision_score(y_true, y_pred_labels, average='macro', zero_division=0)  # 'macro' ou 'weighted' para multiclasse
    recall = recall_score(y_true, y_pred_labels, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred_labels, average='macro', zero_division=0)
    conf_matrix = confusion_matrix(y_true, y_pred_labels)
    
    # AUC-ROC precisa de um one-hot dos rótulos verdadeiros para funcionar em multiclasse
    y_true_binarized = label_binarize(y_true, classes=[0, 1, 2])
    auc = roc_auc_score(y_true_binarized, y_pred, multi_class='ovo')  # ou 'ovr'
    
    # Exibindo os resultados
    print(f'Loss: {x_loss:.4f} - Accuracy: {accuracy:.2f}%')
    print(f'Precisão: {precision:.2f}')
    print(f'Revocação: {recall:.2f}')
    print(f'F1-Score: {f1:.2f}')
    print(f'Matriz de Confusão: \n{conf_matrix}')
    print(f'AUC-ROC: {auc:.2f}')
    print(classification_report(all_labels, all_preds))
```

```python
# Ajustar train_features para a forma correta 
# Função para ajustar o tamanho das sequências
def adjust_sequence(sequence, seq_length):
    if len(sequence) < seq_length:
        # Padding: Preenche com zeros para igualar o comprimento
        padding = np.zeros((seq_length - len(sequence), sequence.shape[1]))  # Mantém o número de features
        return np.vstack((sequence, padding))
    else:
        # Truncamento: Corta as sequências que são maiores do que o seq_length
        return sequence[:seq_length]

# Função de normalização (opcional, se necessário)
def normalize_features(features):
    # Substituir valores NaN ou Inf por zeros, se necessário
    features = np.nan_to_num(features, nan=0.0, posinf=0.0, neginf=0.0)
    
    # Normalização: Subtrai a média e divide pela variância, tratando a variância zero
    mean = np.mean(features, axis=0)
    std = np.std(features, axis=0)
    
    # Evitar divisão por zero (quando std = 0)
    std[std == 0] = 1  # Substitui os valores zero por 1 para evitar divisão por zero
    
    normalized_features = (features - mean) / std
    return normalized_features
```

```python
# Modelo LSTM para Classificação Temporal

class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes, dropout_prob=0.2):
        super(LSTMClassifier, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.dropout = nn.Dropout(dropout_prob)
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        lstm_out = self.dropout(lstm_out)
        final_output = lstm_out[:, -1, :]  # pega a saída do último timestep
        out = self.fc(final_output)
        return out

# Definindo o tamanho de entrada, tamanho oculto e número de classes
# input_size = train_features.shape[1] * 5  # A entrada é achatada, então o número de features por frame * 5 (para cada bbox)
# Atualize o modelo LSTM com o input_size correto
input_size = train_seq_features.shape[2]  # Número de características por frame
hidden_size = 128  # Tamanho do estado oculto, pode ser ajustado
num_classes = 3    # Como você tem 3 classes: "no_fall", "attention" e "fall"

# Criando a instância do modelo LSTM
lstm_model = LSTMClassifier(input_size, hidden_size, num_classes, dropout_prob=0.2)

# Salvar apenas os pesos (recomendado)
# torch.save(lstm_model.state_dict(), 'lstm_model.pth')
# Salvar modelo completo
torch.save(lstm_model, 'lstm_model.pth')
```

```python
# Treinar o LSTM
# Definir critério de perda e o otimizador
weights = torch.tensor([1.0714, 0.5932 , 2.6258])
criterion = nn.CrossEntropyLoss(weight=weights, ignore_index=-1) # Função de loss para classificação
optimizer = optim.Adam(lstm_model.parameters(), lr=0.001) # Otimizador Adam
```

```python
# Treinar o LSTM
epochs = 50

for epoch in range(epochs):
    lstm_model.train()
    running_loss = 0.0

    for batch_features, batch_labels in train_loader:
        # Forward pass
        outputs = lstm_model(batch_features)  # outputs: (batch_size, num_classes)
        
        # Ajustar os labels (substituir -1 por 0, se necessário)
        batch_labels[batch_labels == -1] = 0
        loss = criterion(outputs, batch_labels)  # Usa apenas o último label de cada sequência

        # Backward pass e otimização
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_loader):.4f}")
print(f"Fim")
```

```python
# TESTE COM VIDEO
model_path = "fall_detection_yolo8_person/yolo8_lstm_model7/weights/best.pt"
# model_path = "models/02/best.pt"
# Carregar o modelo treinado
try:
    yolo_model = YOLO(model_path)
    print("Modelo YOLO carregado com sucesso.")
except Exception as e:
    print(f"Erro ao carregar o modelo YOLO: {e}")

# Carregar o modelo completo
try:
    lstm_model = torch.load('lstm_model.pth')
    lstm_model.eval()
    lstm_model.to(device)
    print("Modelo LSTM carregado com sucesso.")
except Exception as e:
    print(f"Erro ao carregar o modelo LSTM: {e}")

# Sequência de features
SEQUENCE_LENGTH = 5
feature_sequence = deque(maxlen=SEQUENCE_LENGTH)

video_path = 'sim/08.mp4'
cap = cv2.VideoCapture(video_path)

# Verifica se abriu
if not cap.isOpened():
    print("Erro ao abrir o vídeo!")
else:
    print("Vídeo carregado com sucesso!")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Erro ao ler o frame!")
        break
    print("Processando frame...")

    # --- 1. YOLOv8 Detection ---
    results = yolo_model.predict(frame, conf=0.5, imgsz=640, verbose=False)
    detections = results[0].boxes.cpu().numpy()  # Get boxes on CPU
    
    # Verifique se há detecções
    print(f"Detecções no frame: {len(detections)}")
    
    # --- 2. Extrair as features do YOLO ---
    frame_features = []
    for box in detections:
        # Exemplo: você pode extrair [x_center, y_center, width, height, conf, class_id]
        xyxy = box.xyxy[0]  # [x1, y1, x2, y2]
        conf = box.conf[0]  # confidence
        cls = box.cls[0]    # class
        x1, y1, x2, y2 = xyxy
        width = x2 - x1
        height = y2 - y1
        x_center = x1 + width / 2
        y_center = y1 + height / 2
    
        # Montar feature para cada detecção (exemplo com 6 features)
        feature = [x_center, y_center, width, height, conf, cls]
        frame_features.extend(feature)
    
    # Verifique quantas características foram extraídas
    print(f"Características extraídas: {len(frame_features)}")
    
    # Preenche com zeros se não houver detecção (garantir entrada do LSTM com mesmo tamanho)
    if len(frame_features) == 0:
        frame_features = [0] * 6  # Supondo que você extrai 6 valores por detecção
    
    # Adiciona as características extraídas à sequência
    feature_sequence.append(frame_features)
    
    # Exibe o conteúdo da sequência
    print(f"Sequência de features antes da predição: {list(feature_sequence)}")
    
    # --- 3. Se tiver sequência suficiente, faz predição com LSTM ---
    if len(feature_sequence) == SEQUENCE_LENGTH:
        # Converter para tensor
        sequence_np = np.array(feature_sequence, dtype=np.float32)
        sequence_tensor = torch.tensor(sequence_np).unsqueeze(0).to(device)
    
        # Fazendo a predição com o LSTM
        with torch.no_grad():
            lstm_output = lstm_model(sequence_tensor)
            prob = torch.softmax(lstm_output, dim=1)
            pred_class = torch.argmax(prob, dim=1).item()
    
        # Labels que você usou
        label_map = {0: 'No Fall', 1: 'Attention', 2: 'Fall'}
        prediction_label = label_map[pred_class]
    
        # Mostrar a predição no vídeo
        cv2.putText(frame, f'{prediction_label} ({prob[0][pred_class]:.2f})',
                    (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    
        # Exibe a predição
        print(f"Predição: {prediction_label} ({prob[0][pred_class]:.2f})")
    
    # --- 4. Mostrar o vídeo ---
    cv2.imshow('Fall Detection YOLO + LSTM', frame)
    
    # Pressione Q para sair
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Verificaçao das classes existentes
# Conjunto para armazenar todas as classes únicas
unique_classes = set()
# Dicionário para contar as ocorrências de cada classe
class_counts = defaultdict(int)

# Função para processar os arquivos de rótulos de cada vídeo
def process_labels(label_dir):
    label_files = sorted(os.listdir(label_dir))
    
    for label_file in label_files:
        label_path = os.path.join(label_dir, label_file)

        # Lê o arquivo de label
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                label_line = f.readline().strip()
                if label_line:  # se a linha não estiver vazia
                    class_id = int(label_line.split()[0])  # pega a classe
                    class_counts[class_id] += 1
                else:
                    class_id = 0  # se o arquivo estiver vazio
                    class_counts[0] += 1 
                unique_classes.add(class_id)
        else:
            unique_classes.add(0)  # Label 0 se não encontrar
            class_counts[0] += 1 

# Processando os arquivos de todos os vídeos
for label_dir in train_video_dirs:
    label_dir = label_dir.replace("images", "labels")
    process_labels(label_dir)

# Resultado
num_classes = len(unique_classes)
print(f"Quantidade de classes únicas: {num_classes}")
print(f"Classes únicas: {sorted(unique_classes)}")

# Verificação de desbalanceamento (VALIDAÇÃO)
total_frames = sum(class_counts.values())
for class_id, count in class_counts.items():
    percentage = (count / total_frames) * 100
    print(f"Classe {class_id}: {count} ocorrências ({percentage:.2f}%)")
```

```python
def validate_model(model, val_loader, criterion):
    model.eval()  # Coloca o modelo em modo de avaliação
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():  # Desativa o cálculo de gradientes para não consumir memória
        for batch_features, batch_labels in val_loader:
            outputs = model(batch_features)  # Predição do modelo

            # Ajuste os labels (substitua -1 por 0)
            batch_labels[batch_labels == -1] = 0

            # Calcular a perda
            loss = criterion(outputs, batch_labels[:, -1])
            running_loss += loss.item()

            # Calcular a precisão
            _, predicted = torch.max(outputs, 1)
            total += batch_labels.size(0)
            correct += (predicted == batch_labels[:, -1]).sum().item()

    avg_loss = running_loss / len(val_loader)
    accuracy = 100 * correct / total
    return avg_loss, accuracy
```

```python
# Treinar o LSTM com validação
epochs = 20

for epoch in range(epochs):
    lstm_model.train()  # Modo de treinamento
    running_loss = 0.0

    for batch_features, batch_labels in train_loader:       
        optimizer.zero_grad()  # Zera os gradientes
        outputs = lstm_model(batch_features)  # Forward pass
        batch_labels[batch_labels == -1] = 0  # Ajuste nos labels

        # Calcular a perda e fazer a retropropagação
        loss = criterion(outputs, batch_labels[:, -1])
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Validação após cada época
    val_loss, val_accuracy = validate_model(lstm_model, val_loader, criterion)

    print(f"Epoch {epoch+1}/{epochs} - Training Loss: {running_loss/len(train_loader):.4f} - Validation Loss: {val_loss:.4f} - Validation Accuracy: {val_accuracy:.2f}%")
```
